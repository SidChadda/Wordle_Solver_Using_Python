### GAME INPUTS:

GREEN SECTION

```python
# Command Line input module

# What I mean by active columns or Green columns is, Which letters become green when you first begin by entering a random letter
df_val_1= pd.DataFrame()
df_val_2= pd.DataFrame()
df_val_3= pd.DataFrame()
df_val_4= pd.DataFrame()
df_val_5= pd.DataFrame()

val1 = input("is the 1st column Green? y/n: ")
if(val1=="y"):
  value_1=input("Enter the letter: ")
  df_val_1=df_new.loc[df_new['letter_1'] == value_1].copy()


val2 = input("is the 2nd column Green? y/n: ")
if(val2=="y"):
  value_2=input("Enter the letter: ")
  df_val_2=df_new.loc[df_new['letter_2'] == value_2].copy()

val3 = input("is the 3rd column Green? y/n: ")
if(val3=="y"):
  value_3=input("Enter the letter: ")
  df_val_3=df_new.loc[df_new['letter_3'] == value_3].copy()

val4 = input("is the 4th column Green? y/n: ")
if(val4=="y"):
  value_4=input("Enter the letter: ")
  df_val_4=df_new.loc[df_new['letter_4'] == value_4].copy()

val5 = input("is the 5th column Green? y/n: ")
if(val5=="y"):
  value_5=input("Enter the letter: ")
  df_val_5=df_new.loc[df_new['letter_5'] == value_5].copy()
```

```python
dataframe_dict={}

if not df_val_1.empty:
  dataframe_dict['letter_1']=df_val_1

if not df_val_2.empty:
  dataframe_dict['letter_2']=df_val_2


if not df_val_3.empty:
  dataframe_dict['letter_3']=df_val_3


if not df_val_4.empty:
  dataframe_dict['letter_4']=df_val_4


if not df_val_5.empty:
  dataframe_dict['letter_5']=df_val_5


```

```python
# Word Combinations generator
int_df=df_new.copy()
for dfy_key in dataframe_dict:
  
  int_df = pd.merge(int_df, dataframe_dict[dfy_key], how ='inner', indicator=False)

print("\n")
print("All the possible 5 Letter word combinations are: ")
int_df
```

FINE TUNE & YELLOW SECTION:

```python
#Run the below script if you want to look up the Yellow category letter - word combinations
#Yellow Letter meaning- the Letter is correct but placed in incorrect position


df_lookup_letter=int_df.copy()

letter_looker = input("Do you want to look up a certain letter? y/n:  ")
if(letter_looker=="y"):
  letter_looker_value=input("Enter the list of letters you want to search (seperate the letters by , (it's a comma, please no space)): ")
  letter_looker_list=letter_looker_value.split(",")


  for lookup_letter in letter_looker_list:
    df_lookup_letter=df_lookup_letter[df_lookup_letter.isin([lookup_letter]).any(axis=1)].copy()

df_final = df_lookup_letter.copy()
df_final = df_final.reset_index()
df_final=df_final.drop(['index'], axis = 1)


print("\n")
print("All the possible 5 Letter word combinations are: ")
df_final
```
